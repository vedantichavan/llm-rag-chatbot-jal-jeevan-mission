```python
%%writefile app.py
import streamlit as st
from transformers import pipeline

# Load the pre-trained question-answering model from HuggingFace
qa_pipeline = pipeline("question-answering", model="distilbert-base-cased-distilled-squad")

# Sample Jal Jeevan Mission related knowledge
jjm_knowledge = """
    The **Jal Jeevan Mission (JJM)** is a flagship program launched by the **Government of India** aimed at providing safe and adequate drinking water to rural households through individual household tap connections by 2024.

    Here’s detailed information on the **Jal Jeevan Mission**:

    ### 1. **How the Jal Jeevan Mission Started**
      - **Launch Date**: The Jal Jeevan Mission was launched on **August 15, 2019** by **Prime Minister Narendra Modi**.
      - **Objective**: The mission aims to provide **tap water connections** to **every rural household** in India by **2024**. It emphasizes sustainability, water conservation, and community participation.
      - **Government Vision**: It was part of the **Ministry of Jal Shakti's** initiative under the **National Rural Drinking Water Program (NRDWP)** to ensure **safe drinking water** is available to rural households.

    ### 2. **Who Started the Jal Jeevan Mission**
      - **Prime Minister Narendra Modi** formally launched the mission.
      - **Ministry of Jal Shakti**: The program is being implemented by the Ministry of Jal Shakti, with the Department of Drinking Water and Sanitation (DDWS) coordinating the efforts.
      - **Collaboration with States**: The implementation is carried out in partnership with state governments and local bodies, allowing states to adapt the program based on local needs.

    ### 3. **What the Jal Jeevan Mission Does**
      The key focus areas of the **Jal Jeevan Mission** are:
      - **Providing Tap Water Connections**: The mission aims to provide **safe and sufficient drinking water** to every rural household through piped water connections.
      - **Water Quality Monitoring**: Ensuring water quality is maintained and making sure that **safe water** is provided to all households.
      - **Community Involvement**: The mission emphasizes **community ownership**, with local bodies like **Gram Panchayats** playing a central role in water management.
      - **Sustainability**: Focus on **rainwater harvesting**, **groundwater recharge**, and **water use efficiency**.
      - **Promotion of Local Solutions**: In rural areas where piped water is not feasible, alternative solutions like **rainwater harvesting**, **water storage tanks**, and **water ATMs** are being explored.

    ### 4. **Statistical Reports and Key Data**
      - **Number of Tap Connections**: As of 2023, more than **10 crore** (100 million) tap water connections have been provided across rural India.
      - **Impact on Rural Households**: The mission has connected millions of households, significantly improving access to clean drinking water, reducing dependency on unsafe water sources like handpumps and open wells.
      - **Funding and Investment**: The government has allocated over **₹3.6 lakh crore** (about **USD 48 billion**) for the mission over the 5-year period (2019-2024).

    ### 5. **States with the Most Impact**
      Some states and Union Territories have made significant progress under JJM. Here are a few examples:
      - **Goa**: Goa was one of the first states to achieve 100% tap water connections to rural households.
      - **Himachal Pradesh**: This state has also made rapid strides in providing tap water to its rural areas.
      - **Telangana**: Known for its significant increase in tap water coverage in rural areas.
      - **Uttarakhand**: Uttarakhand has seen remarkable progress in rural water supply as part of JJM.

    ### 6. **Before and After: Impact of the Mission**
      - **Before JJM (Pre-2019)**: Many rural areas in India were still dependent on unsafe sources like handpumps, wells, and open water bodies. Around **17% of rural households** did not have access to piped water, and women and children often had to spend several hours each day collecting water.
      - **After JJM (Post-2019)**: By 2023, more than **50% of rural households** had access to piped water connections, and the mission is expected to meet its target of providing **tap water to every household** by **2024**.

    ### 7. **Percentage of Rural Households with Tap Water Connections (Before vs After)**
      - **Before JJM**: Around **17%** of rural households had access to tap water connections in 2019.
      - **After JJM**: As of 2023, more than **50%** of rural households have been provided with tap water connections.

    ### 8. **Challenges and Solutions**
      - **Challenges**:
        - In many areas, poor water quality and scarcity of water resources have hindered progress.
        - Infrastructure challenges, particularly in remote and hilly areas, are significant.
      - **Solutions**:
        - Focus on water conservation through community participation.
        - Introducing **water ATMs** and **local water management initiatives**.
        - Use of **solar-powered pumps** and **rainwater harvesting** systems.

    ### Summary of Key Points:
    - **Launched**: August 15, 2019, by Prime Minister Narendra Modi.
    - **Aim**: To provide **tap water to all rural households** by 2024.
    - **Implementation**: Led by the Ministry of Jal Shakti with local bodies' participation.
    - **Impact**: Over **10 crore (100 million)** connections, with rural tap water coverage increasing from **17%** to **50%**.
    - **States Leading**: Goa, Himachal Pradesh, Telangana, and Uttarakhand.

      1. Coverage of Tap Water Connections
    Total Households in Rural India (as of 2021): Approximately 18.8 crore (188 million) rural households.
    Tap Water Coverage in 2019: Before the launch of the Jal Jeevan Mission, only 17% of rural households had access to tap water.
    Current Tap Water Coverage (2023): Over 10 crore (100 million) households have been provided with tap water connections, bringing the overall rural tap water coverage to over 50%.
    Target: To achieve 100% coverage of rural households by 2024, i.e., all 18.8 crore rural households.
    2. Progress by State/UT
    Here’s the impact of Jal Jeevan Mission across different states:

    Top-performing States (as of 2023)
    Goa:
    100% Coverage: Goa became the first state to provide tap water to all rural households.
    Himachal Pradesh:
    96.5% Coverage: This state has made remarkable progress, with over 96% of rural households now having tap water connections.
    Telangana:
    91.5% Coverage: One of the leading states in terms of providing water connections. Almost 92% of rural households now have tap water.
    Uttarakhand:
    87% Coverage: A significant increase in water connections, with almost 9 out of 10 rural households having access to piped water.
    Andhra Pradesh:
    83% Coverage: This state has also seen a significant rise in rural tap water coverage.
    States with High Progress
    Bihar: Achieved 62% coverage, which is notable for a state with a large rural population.
    Rajasthan: 50% of its rural households now have tap water connections.
    Madhya Pradesh: 47% coverage of tap water connections in rural households.
    States with Low Coverage (Challenges)
    Uttar Pradesh: Although it is making progress, it has about 34% coverage. The state's large rural population presents challenges in achieving full coverage.
    Jharkhand: The state has around 38% coverage. Water quality and infrastructure challenges are significant.
    3. Financial Investment
    Total Allocation (2019-2024): The Government of India allocated over ₹3.6 lakh crore (approximately USD 48 billion) for the implementation of the Jal Jeevan Mission.
    State-wise Funding: Each state has received funds based on its rural population and current water coverage. States with greater coverage challenges receive a larger share of the funding.
    Per Household Cost: On average, it costs about ₹30,000 to provide a tap water connection to a household. This includes the cost of infrastructure, water source development, and ensuring water quality.
    4. Water Quality and Sustainability Initiatives
    Water Testing: Over 50 lakh water samples are tested annually to ensure that the water supplied is safe and potable.
    Community Participation: The mission focuses on empowering Gram Panchayats (village councils) and local communities to take ownership of water management. This helps in sustainability and better resource management.
    Water Conservation: The mission is promoting rainwater harvesting and groundwater recharge as part of its strategy to ensure long-term sustainability.
    5. Rural Drinking Water Schemes and Alternative Solutions
    Rainwater Harvesting: About 50% of villages are implementing rainwater harvesting techniques to enhance local water storage and reduce dependency on external sources.
    Water ATMs: In areas where piped water is not feasible, water ATMs are being set up in rural areas to provide clean drinking water at affordable rates.
    Solar-Powered Water Pumps: In regions with limited electricity access, solar-powered pumps are being used to draw water, ensuring sustainability in remote areas.
    6. Women and Community Empowerment
    Women’s Role: Women, who traditionally spend a significant amount of time fetching water, benefit the most from the mission. Having water available at home saves time and allows them to focus on other activities such as education, health, and livelihood.
    Gram Panchayats' Involvement: Local governance bodies play a central role in water management, which helps to ensure long-term community ownership and participation.
    7. Health and Socio-Economic Benefits
    Health Impact: Access to safe drinking water reduces the incidence of waterborne diseases, such as diarrhea, cholera, and typhoid, especially in rural areas.
    Economic Benefits: Improved water supply contributes to better productivity, particularly in rural sectors like agriculture, education, and women’s employment.
    8. Expected Benefits by 2024
    100% Coverage of Rural Households: By 2024, it is expected that all rural households will have access to tap water, significantly improving their quality of life.
    Better Water Management: The mission will lead to better water conservation, recycling, and use of alternative water sources like rainwater harvesting and water recycling plants.
    Reduction in Health Costs: The reduction in waterborne diseases will lessen healthcare burdens in rural communities.
    9. Statistical Highlights (2023)
    Total Number of Rural Households: ~188 million (18.8 crore).
    Households Covered: ~10 crore (100 million) households.
    Funds Allocated: ₹3.6 lakh crore (USD 48 billion).
    States with 100% Coverage: Goa (first state to achieve 100%).
    States with ≥90% Coverage: Himachal Pradesh, Telangana, and Haryana.
    National Coverage: ~50% of rural households covered by tap water.
    10. Comparison (Before and After JJM)
    Before JJM (2019):
    Tap water coverage in rural areas: 17%.
    Lack of clean water was a major cause of diseases.
    Rural women spent hours fetching water.
    After JJM (2023):
    Tap water coverage in rural areas: 50% (with 100% expected by 2024).
    Major improvement in public health with reduced waterborne diseases.
    Empowerment of women and increased focus on education and work.
"""

# Function to get the answer to the question
def get_answer(query):
    result = qa_pipeline({
        'question': query,
        'context': jjm_knowledge
    })
    return result['answer']

# Streamlit app title
st.title("Jal Jeevan Mission - Q&A")

# Input field for user query
user_query = st.text_input("Ask a question about Jal Jeevan Mission:")

# Button to trigger the response
if st.button("Get Answer"):
    if user_query:
        answer = get_answer(user_query)
        st.write(f"Answer: {answer}")
    else:
        st.write("Please enter a question.")
```
